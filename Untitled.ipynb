```julia
# data = CSV.read("car data.csv")
data = DelimitedFiles.readdlm("car data.csv",',')
data = data[2:size(data,1),:]
data
```

```julia
#data ditransformasi, string seperti nama terlalu variant jadi diabaikan
data = data[:,2:size(data,2)]
for i in 1:size(data,1)
    #data tahun karena nilai tahun tidak relevan, diubah menjadi tahun ini(tahun data) kurang tahun keluar
    data[i,1] = 2018 - data[i,1]
    #data bensin, menurut riset, diesel lebih efisien dari petrol, jadi diesel dianggap 2, petrol 1
    if data[i,5] == "Petrol"
        data[i,5] = 1
    else
        data[i,5] = 2
    end
    #karena dealer lebih terjamin daripada penjualan individu, dealer dianggap 2, individu 1
    if data[i,6] == "Dealer"
        data[i,6] = 2
    else
        data[i,6] = 1
    end
    #karena Matic lebih mahal dari manual, bobotnya terhadap harga lebih berat, jadi matic 2 manual 1
    if data[i,7] == "Automatic"
        data[i,7] = 2
    else
        data[i,7] = 1
    end
    #karena tidak tiap km dianggap dalam km driven, kita div dengan 2000 dengan asumsi perbedaan 2000 km baru bisa dianggap beda
    data[i,4] = data[i,4] / 2000
    #karena how my program works, yang menjadi Y akan ditaruh, 'ditukar' dengan paling ujung kiri
    temp = data[i,3]
    data[i,3] = data[i,8]
    data[i,8] = temp
end
# data_floated = Array{Float64,2}(undef,size(data,1),size(data,2))
# for i in 1:size(data,1)
#     for j in 1:size(data,2)
#         data_floated
#     end
# end
data = convert(Array{Float64,2},data)

# data = data[1:6,:]
data
    
```

```julia
include("linearreg.jl")
using .LinReg
using LinearAlgebra

data_train = data[1:271,:]
#dimensix, dimensiy, learning_rate 
Regressor = LinReg.linreg(7,1,0.0005)
# Regressor.weight
LinReg.train(Regressor,data_train,10000)
Regressor.weight
```

```julia
#plotting y terhadap umur mobil
result = zeros(30,1)
features = data[272:301,1:size(data,2)-1]
labels = data[272:301,size(data,2)]
# dot(features[1,:],Regressor.weight)
for i in 1:30
    result[i,1] = dot(features[i,:],Regressor.weight)
end

factor1 = ones(30)
year = data[272:301,1]
yp = [factor1 year]
coYear = yp\result
f1 = coYear[1]*factor1 + coYear[2]*year
# year[:,1] = 2018 .- year[:,1]

p1 = scatter(year, labels, color=:green, label="Actual")
p1 = plot!(year, f1, linewidth=1, color=:red, label = "Prediction")

plot(p1)
```

```julia
MSE = 0
RMSE = 0
MAE = 0
for i in 1:30
    MSE += (labels[i] - result[i])^2
    MAE += abs(labels[i] - result[i])
end
MSE = MSE/30
MAE = MAE/30
RMSE = sqrt(MSE)
print("MSE ")
print(MSE)
print("\nMAE ")
print(MAE)
print("\nRMSE ")
print(RMSE)
```
