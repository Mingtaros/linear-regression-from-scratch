## Prediksi Harga Jual Mobil dengan Regresi Linear
Dataset untuk regresi linear ini diunduh dari https://www.kaggle.com/nehalbirla/vehicle-dataset-from-cardekho.
Pada kasus ini, tiga variabel independen yang dipilih dalam melakukan prediksi adalah tahun ('Year'), jumlah kilometer yang sudah dilalui mobil ('Kms_driven') dan harga terkini mobil ('Present_Price'). Ketiga aspek ini dapat dimanfaatkan untuk memprediksi harga jual mobil ('Selling_Price') sebagai variabel dependen dengan memanfaatkan multiple linear regression.

```python
import linreg
```

## Membaca File csv

```python
df = linreg.read_csv("car_data.csv")
df
```

## Mengambil Data Numerik
Dalam hal ini, data yang kita akan manfaatkan berasal dari kolom 'Year', 'Present_Price', 'Kms_Driven', 'Selling_Price'

```python
used_data = df[['Year','Present_Price', 'Kms_Driven', 'Selling_Price']]
used_data
```

## Melakukan Splitting Data
Dataset akan displit secara horizontal menjadi data untuk train dan test.

```python
train, test = linreg.split_train_test(used_data)
train
```

```python
test
```

## Melakukan Training Data
Data yang dialokasikan untuk training akan dimanfaatkan untuk menghitung koefisien dan konstanta pada model regresi yang kita rumuskan. Dalam hal ini, variabel independen akan dikelompokkan dalam matriks x_var dan variable dependen akan dikelompokkan dalam matriks y_var.

```python
x_var, y_var = linreg.split_variables(train)
coeffs = linreg.calculate_const_coefs(x_var, y_var)
coeffs
```

Fungsi calculate_const_coefs menghasilkan konstanta dan koefisien pada persamaan regresi. Dalam hal ini, persamaan regresi yang berhasil dihasilkan adalah:
**y = -971.3702 + 0.4828x1 + 0.5146x2 - 1.3608e-07 x3**
dengan x1, x2, dan x3, merupakan Year, Present_Price, dan Kms_Driven secara berurutan. 

## Melakukan Prediksi
Prediksi dilakukan dengan menggunakan data test yang dimasukkan ke persamaan regresi yang berhasil dirumuskan. Hasilnya kemudian akan dibandingkan relatif dengan actual result nya dengan menghitung nilai **RSquared** nya.

```python
x_test, y_test = linreg.split_variables(test)
pred = linreg.predict(x_test, coeffs)
rsq = linreg.calculate_rsquared(y_test, pred)
rsq
```

Nilai RSquared 0.8476 menunjukkan Quality of Fit yang dimiliki model regresi kita. Semakin dekat nilainya ke satu, maka model regresi dapat dikatakan semakin merepresentasikan keadaan sesungguhnya.

Membandingkan actual value 'Selling_Price' dengan predicted value 'Prediction'.

```python
test['Prediction'] = pred
test
```

## Visualisasi Data
Hasil regresi dapat kita bandingkan secara visual dengan menampilkan scatter plot serta garis hasil regresi linear.

```python
import numpy as np
x_axis = x_test.transpose()
y_axis = np.array(y_test.transpose()).flatten()
linreg.draw_regression(x_axis, y_axis, coeffs, 1, "Year")
```

```python
linreg.draw_regression(x_axis, y_axis, coeffs, 2, "Present_Price")
```
